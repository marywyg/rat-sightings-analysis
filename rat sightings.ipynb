1.	Utworzenie zbioru danych:

```python
import pandas as pd

df = pd.read_csv("Rat_Sightings.csv")
df
```

2.	Podstawowe statystyki danych:

```python
# Podstawowe statystyki danych dla df

print("Wymiary danych (wiersze, kolumny):", df.shape)
print("\nTypy danych:")
print(df.dtypes)

print("\nStatystyki opisowe (kolumny numeryczne):")
display(df.describe())

print("\nStatystyki opisowe (kolumny kategoryczne):")
display(df.describe(include="object"))

print("\nBraki danych (top 20 kolumn):")
display(df.isna().sum().sort_values(ascending=False).head(20))
```

3.	Analiza braków danych, przypadków ostających, ekstremalnych. Wpływ przypadków na statystyki.

```python
# 3. Analiza braków danych, przypadków odstających i ekstremalnych + wpływ na statystyki

print("=== 1) Analiza braków danych ===")
missing = df.isna().sum().to_frame("braki")
missing["% braków"] = (missing["braki"] / len(df) * 100).round(2)
missing = missing.sort_values(["braki", "% braków"], ascending=False)

display(missing.head(25))

kolumny_puste = missing[missing["braki"] == len(df)].index.tolist()
print(f"\nKolumny całkowicie puste ({len(kolumny_puste)}): {kolumny_puste}")

print("\n=== 2) Analiza przypadków odstających (IQR) w kolumnach numerycznych ===")
num_df = df.select_dtypes(include="number")

outlier_rows = []
for col in num_df.columns:
    s = num_df[col].dropna()
    if len(s) < 30 or s.nunique() < 5:
        continue

    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1

    if iqr == 0:
        continue

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    outliers = s[(s < lower) | (s > upper)]
    outlier_rows.append({
        "kolumna": col,
        "N": len(s),
        "Q1": q1,
        "Q3": q3,
        "IQR": iqr,
        "dolna_granica": lower,
        "górna_granica": upper,
        "liczba_odstających": len(outliers),
        "%_odstających": round(len(outliers) / len(s) * 100, 2),
        "min": s.min(),
        "p1": s.quantile(0.01),
        "p99": s.quantile(0.99),
        "max": s.max(),
    })

outliers_df = pd.DataFrame(outlier_rows).sort_values("%_odstających", ascending=False)
display(outliers_df)

print("\n=== 3) Wpływ wartości odstających na statystyki (przed vs po odcięciu IQR) ===")
impact_rows = []
for _, r in outliers_df.iterrows():
    col = r["kolumna"]
    s = num_df[col].dropna()
    lower = r["dolna_granica"]
    upper = r["górna_granica"]

    s_trim = s[(s >= lower) & (s <= upper)]
    if len(s_trim) == 0:
        continue

    mean_before = s.mean()
    mean_after = s_trim.mean()
    std_before = s.std()
    std_after = s_trim.std()
    med_before = s.median()
    med_after = s_trim.median()

    mean_change_pct = ((mean_after - mean_before) / mean_before * 100) if mean_before != 0 else None
    std_change_pct = ((std_after - std_before) / std_before * 100) if std_before != 0 else None

    impact_rows.append({
        "kolumna": col,
        "średnia_przed": mean_before,
        "średnia_po": mean_after,
        "zmiana_średniej_%": None if mean_change_pct is None else round(mean_change_pct, 2),
        "mediana_przed": med_before,
        "mediana_po": med_after,
        "odch_std_przed": std_before,
        "odch_std_po": std_after,
        "zmiana_odch_std_%": None if std_change_pct is None else round(std_change_pct, 2),
        "usunięte_obs": int(len(s) - len(s_trim)),
    })

impact_df = pd.DataFrame(impact_rows)

if not impact_df.empty:
    impact_df = impact_df.sort_values(
        by="zmiana_średniej_%",
        key=lambda x: x.abs() if x.notna().any() else x,
        ascending=False
    )
    display(impact_df)
else:
    print("Brak kolumn numerycznych spełniających warunki analizy wpływu.")

print("\n=== 4) Krótkie wnioski automatyczne ===")
if len(kolumny_puste) > 0:
    print(f"- W danych występują kolumny całkowicie puste: {len(kolumny_puste)} (warto rozważyć usunięcie).")
else:
    print("- Brak kolumn całkowicie pustych.")

if not outliers_df.empty:
    top_out = outliers_df.iloc[0]
    print(f"- Najwyższy udział wartości odstających: {top_out['kolumna']} ({top_out['%_odstających']}%).")
else:
    print("- Nie wykryto wartości odstających metodą IQR (dla analizowanych kolumn).")

if not impact_df.empty:
    top_impact = impact_df.iloc[0]
    print(
        f"- Największy wpływ na średnią po odcięciu odstających: {top_impact['kolumna']} "
        f"(zmiana {top_impact['zmiana_średniej_%']}%)."
    )
else:
    print("- Nie wyznaczono wpływu na statystyki.")
```

```python
# Usunięcie całkowicie pustych kolumn
df = df.drop(columns=kolumny_puste)

print(f"Usunięto {len(kolumny_puste)} całkowicie pustych kolumn.")
print("Nowy rozmiar danych:", df.shape)

df
```

4.	Wykresy ramka – wąsy

```python
import seaborn as sns
import calendar

import matplotlib.pyplot as plt

# ensure datetime columns
df['Created Date'] = pd.to_datetime(df['Created Date'], errors='coerce')
df['Closed Date'] = pd.to_datetime(df['Closed Date'], errors='coerce')

# 1) Czas rozwiązania sprawy (dni) vs Typ lokalizacji
df['resolution_days'] = (df['Closed Date'] - df['Created Date']).dt.total_seconds() / 86400.0
res_df = df.dropna(subset=['resolution_days', 'Location Type'])
res_df_vis = res_df[res_df['resolution_days'] >= 0].copy()
y_max = res_df_vis['resolution_days'].quantile(0.95)

plt.figure(figsize=(12,6))
sns.boxplot(data=res_df_vis, x='Location Type', y='resolution_days')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Typ lokalizacji')
plt.ylabel('Czas rozwiązania (dni)')
plt.title('Czas rozwiązania sprawy a typ lokalizacji')
if pd.notna(y_max) and y_max > 0:
    plt.ylim(0, y_max)
plt.tight_layout()
plt.show()

# Przygotowanie dziennej liczby zgłoszeń (używane w wykresach 2 i 3)
created = df.dropna(subset=['Created Date']).copy()
daily_counts = created.groupby(created['Created Date'].dt.date).size().rename('count').to_frame()
daily_counts['date'] = pd.to_datetime(daily_counts.index)
daily_counts['weekday'] = daily_counts['date'].dt.day_name()
daily_counts['month_name'] = daily_counts['date'].dt.month_name()

# 2) Zgłoszenia a dzień tygodnia (boxplot na dziennych liczbach rozłożonych po dniach tygodnia)
weekday_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
# przetłumaczenie nagłówków na polski (opcjonalne)
weekday_labels_pl = ['Poniedziałek','Wtorek','Środa','Czwartek','Piątek','Sobota','Niedziela']

plt.figure(figsize=(10,5))
sns.boxplot(data=daily_counts, x='weekday', y='count', order=weekday_order)
plt.xticks(ticks=range(7), labels=weekday_labels_pl, rotation=0)
plt.xlabel('Dzień tygodnia')
plt.ylabel('Liczba zgłoszeń (dziennie)')
plt.title('Zgłoszenia a dzień tygodnia')
plt.tight_layout()
plt.show()

# 3) Dzienna liczba zgłoszeń a miesiąc w roku (boxplot)
month_order = list(calendar.month_name)[1:]  # January..December
month_labels_pl = ['Sty','Lut','Mar','Kwi','Maj','Cze','Lip','Sie','Wrz','Paź','Lis','Gru']

plt.figure(figsize=(12,5))
sns.boxplot(data=daily_counts, x='month_name', y='count', order=month_order)
plt.xticks(ticks=range(12), labels=month_labels_pl, rotation=45, ha='right')
plt.xlabel('Miesiąc')
plt.ylabel('Dzienna liczba zgłoszeń')
plt.title('Dzienna liczba zgłoszeń a miesiąc w roku')
plt.tight_layout()
plt.show()
```
